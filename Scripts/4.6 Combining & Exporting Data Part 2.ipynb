```python
# Import Libraries
import pandas as pd
import numpy as np
import os
```

```python
# Define Path
path = r'C:\Users\cshat\06-2023 Instacart Basket Analysis'
```

```python
# Import Dataframes
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
df_ords_prods_combined = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
# Check Dataframes
df_ords_prods_combined.shape
```

```python
df_ords_prods_combined.head()
```

```python
df_prods.head()
```

```python
# Drop _merge Column
df_ords_prods_combined = df_ords_prods_combined.drop(columns = ['_merge'])
```

```python
# Merge Dataframes
df_merged = df_ords_prods_combined.merge(df_prods, on = 'product_id', indicator = True)
```

```python
df_merged.head()
```

```python
# Drop Columns
df_merged = df_merged.drop(columns = ['Unnamed: 0.1','Unnamed: 0_x'])
```

```python
# Export Data
df_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
