```python
# Create Flags
df_orders_products_merged['max_order'] = df_orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Derive Columns
df_orders_products_merged.loc[df_orders_products_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_orders_products_merged.loc[(df_orders_products_merged['max_order'] <= 40) & (df_orders_products_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_orders_products_merged.loc[df_orders_products_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

New customers have the highest mean of purchase prices, however there isn't much variation between the loyalty groups.

```python
# Step 6
df_orders_products_merged['mean_price'] = df_orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_orders_products_merged.head()
```

```python
df_orders_products_merged.loc[df_orders_products_merged['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_orders_products_merged.loc[df_orders_products_merged['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_orders_products_merged.head()
```

```python
df_orders_products_merged['spending_flag'].value_counts(dropna = False)
```

```python
# Step 7
df_orders_products_merged['median_days'] = df_orders_products_merged.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
df_orders_products_merged.head()
```

```python
df_orders_products_merged.loc[df_orders_products_merged['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
df_orders_products_merged.loc[(df_orders_products_merged['median_days'] > 10) & (df_orders_products_merged['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_orders_products_merged.loc[df_orders_products_merged['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
df_orders_products_merged.head()
```

```python
df_orders_products_merged['frequency_flag'].value_counts(dropna=False)
```

```python
# Assessing Null Values
print(df_orders_products_merged[df_orders_products_merged.frequency_flag.isnull()])
```

```python
df_orders_products_merged['frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
df_orders_products_merged['frequency_flag'].value_counts(dropna=False)
```

Null values are due to customers only having made one order. Therefore, those entries were replaced with 'Non-frequent customer'.
